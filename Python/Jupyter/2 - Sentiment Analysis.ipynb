```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download necessary resources
nltk.download('vader_lexicon')

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Dessertation/processed_bluesky_data.csv", encoding="latin1")

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Apply VADER sentiment analysis
df['sentiment_score'] = df['text'].fillna("").apply(lambda text: sia.polarity_scores(text)['compound'])

# Classify sentiment based on score
def classify_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df['sentiment_label'] = df['sentiment_score'].apply(classify_sentiment)

# Save results to CSV
df.to_csv("/content/drive/MyDrive/Dessertation/sentiment_results.csv", index=False)

# Show some results
print(df[['text', 'sentiment_score', 'sentiment_label']].head())
```
